```python
#%%Ejercicio 9
x1 = float(input('ingrese x1 -> ')) #Datos de entrada
y1 = float(input('ingrese y1 -> '))
r1 = float(input('ingrese r1 -> '))
x2 = float(input('ingrese x2 -> '))
y2 = float(input('ingrese y2 -> '))
r2 = float(input('ingrese r2 -> '))

a = float(input('ingrese cordenada del punto en a del eje x -> '))
b = float(input('ingrese cordenada del punto en b del eje y -> '))

xpri=abs(x1-a) #Dista de coordenada del eje x con coorde del punto de circu 1
ypri=abs(y1-b) #Dista de coordenada del eje y con coorde del punto de circu 1

if r1>=xpri and r1>=ypri: #si radio es mayor a dista entre coordenadas haga
    print ('El punto esta dentro del primer circulo') #Dato Salida
    c1=True #Guardar verdadero si esta dentro del 1 circulo
else:
    print ('El punto NO esta dentro del primer circulo') #Dato Salida
    c1=False #Guardar falsoo si esta dentro del 1 circulo

xseg=abs(x2-a) #Resta de coordenada del eje x con coorde del punto de circu 2
yseg=abs(y2-b) #Resta de coordenada del eje y con coorde del punto de circu 2

if r2>=xseg and r2>=yseg: #si radio es mayor a dista entre coordenadas haga
    print ('El punto esta dentro del segundo circulo') #Dato Salida
    c2= True#Guardar verdadero si esta dentro del 2 circulo
else:
    print ('El punto NO esta dentro del segundo circulo') #Dato Salida
    c2=False #Guardar Falso si esta dentro del 2 circulo

#Evaluar verdaderos y falsos y dar datos de salida
if c1==True and c2==True: 
    print('El punto esta dentro de ambos circulos') #Dato Salida
else:
    print('El punto NO esta dentro de ambos circulos') #Dato Salida

if c1==False and c2==False:
    print('El punto esta fuera de ambos circulos') #Dato Salida
else:
    print('El punto NO esta fuera de ambos circulos') #Dato Salida
 
```

```python
#%% Ejercicio 15
o= int( input ('Ingrese el numero de datos que tiene la lista -> ')) #entrada
i=0 #declaro lista de elementos ingresados y i=0 para ciclo
lis=[]
while i < o: #ejecute las veces que se indico anterior
    l=str(input('Ingrese numero de la lista en orden -> ')) #Datos entrada
    lis.append(l) #agrego a lista
    i=i+1
repe=[] #se crea lista repe
r=0 #declaro variable r para contador
for e in lis: #recorre lista
    if e not in repe: #si variable que recorre no esta en repe haga
        rep=lis.count(e) #cuenta cuantas veces esta la letra en lis
        if rep==2: #si contador =2, haga
            r=r+1 #sume al contador
            repe.append(e)     #agregue a lista de repe 2 veces
print(f'\n{r} son los elementos que se repiten exactamente dos veces') #Salida
```
